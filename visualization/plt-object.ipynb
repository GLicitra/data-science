## Plot figure via object oriented

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a dataset
x = np.linspace(start = 0, stop = 10, num=50, endpoint=True)
y = np.linspace(start = 1, stop = 5, num=50, endpoint=True)
```

**Example 1**

```python
# Create figure and axis objects using subplots()
fig, axis = plt.subplots()
# prepare initial stype option
options = dict(marker = 'o', color = 'blue')
# call the plot method on the axis object
line = axis.plot(x, y, **options)
# modify axis object with the set methods
_ = axis.set_ylabel('Times')
_ = axis.set_xlabel('Distance')
# displace figure
plt.show()
```

**Example 2**

```python
# Create figure and axis objects using subplots()
fig, axis = plt.subplots()
# Plot line using the axis.plot() method
line = axis.plot(x, y, linestyle=" ", marker="o", color="red")
# Use the plt.show() method to display the figure
plt.show()
```
